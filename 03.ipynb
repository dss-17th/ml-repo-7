```python
data = data[data['c6']==0] # c6 법인이용 개인 고객 = 0 / 법인 = 1 / 법인 구성원 = 2
data = data[(data['c7']==0) | (data['c7']==40) | (data['c7']==100)] # c7 과실율 unique = 100,   0,  90,  55,  30,  80,  40,  10,  70,  20,  60,  85,  75, 50,  15,  65,  25
data = data[data['c12']!=5] # c12 대여기간 (2,3,4,5) : 1 / (6,7,8,9) : 2 / (10 =< x <=36) : 3 / (36+) : 4 / (0,1) : 5
data = data[(data['c19']==1) | (data['c19']==2) | (data['c19']==3)] # c19 수리 부위 갯수
data['c25'] = data['c15']-data['c14'] # c25 : 보험사 손해비용 - 차량수리비용
```

```python
data
```

```python
def outliers_iqr(data):
    q1, q3 = np.percentile(data, [25,75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    
    return np.where((data > upper_bound)|(data < lower_bound))
```

```python
c25_outlier_index = outliers_iqr(data['c25'])[0]
```

```python
c25_not_outlier_index = []

for i in data.index:
    if i not in c25_outlier_index:
        c25_not_outlier_index.append(i)
```

```python
c25_outlier_index
```

```python
c25_not_outlier_index = np.array(c25_not_outlier_index)
```

```python
data = data.loc[c25_not_outlier_index]
```

```python
len(data[data["fraud"] == 1])
```

```python
final = pd.concat([data, data1])
final = final.drop(['c25'], 1)
```

```python
final
```

```python
len(final[final["fraud"] == 1])
```

```python
data = final
```

## 언더 샘플링

### RandomUnderSampler

```python
X = data.drop("fraud", 1)
y = data["fraud"]
```

```python
from imblearn.under_sampling import RandomUnderSampler 
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)
```

```python
#from collections import Counter
#print('Resampled dataset shape %s' % Counter(y))
```

```python
from sklearn.model_selection import train_test_split
 
X = X_res
y = y_res
 
# 테스트 데이터 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y)
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics    
 
# 학습 진행
forest = RandomForestClassifier(n_estimators=100)
forest.fit(X_train, y_train)
 
# 예측
y_pred = forest.predict(X_test)
print(y_pred)
print(list(y_test))
 
# 정확도 확인
print('정확도 :', metrics.accuracy_score(y_test, y_pred))

print('precision', metrics.precision_score(y_test, y_pred))

print('recall', metrics.recall_score(y_test, y_pred))

print('f1', metrics.f1_score(y_test, y_pred))
```
